**Removing Outliers**

```python
data= df.dropna()
```

```python
def remove_outliers(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data
```

```python
columns_to_check = ['x.radius_mean','x.texture_mean','x.perimeter_mean'	,'x.area_mean','x.smoothness_mean','x.compactness_mean',
	'x.concavity_mean'	,'x.concave_pts_mean'	,'x.symmetry_mean',	'x.fractal_dim_mean','x.texture_worst',	'x.perimeter_worst',
  	'x.area_worst',	'x.smoothness_worst',	'x.compactness_worst',	'x.concavity_worst'	,'x.concave_pts_worst',	'x.symmetry_worst',	'x.fractal_dim_worst']
```

```python
data_cleaned = remove_outliers(data, columns_to_check)
```

```python
plt.figure(figsize=(18,20))
for i,col in enumerate(data_cleaned.columns):
  plt.subplot(6,6,i+1)
  sns.boxplot(x=data_cleaned[col])
  plt.title(col)
plt.tight_layout()
plt.show()
```
