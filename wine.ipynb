**Removing Outliers**

```python
data_cleaned= data.dropna()
```

```python
def remove_outliers(data_cleaned, columns):
    for col in columns:
        Q1 = data_cleaned[col].quantile(0.25)
        Q3 = data_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data_cleaned = data_cleaned[(data_cleaned[col] >= lower_bound) & (data_cleaned[col] <= upper_bound)]
    return data_cleaned
```

```python
columns_to_check = ['Malic_acid', 'Ash', 'Alcalinity_of_ash', 'Magnesium', 'Proanthocyanins', 'Color_intensity', 'Hue']
```

```python
data_cleaned1 = remove_outliers(data_cleaned, columns_to_check)
```

```python
plt.figure(figsize=(18,20))
for i,col in enumerate(data_cleaned1.columns):
  plt.subplot(4,4,i+1)
  sns.boxplot(x=data_cleaned1[col])
  plt.title(col)
plt.tight_layout()
plt.show()
```

**Correlation Analysis**

```python
data_cleaned1.corr()
```

Form Above we can observe

Strong Positive Correlation: For example, Flavanoids and Total_phenols have a correlation coefficient of 0.882622. This indicates a strong positive relationship between these two variables. Wines with higher flavanoid content tend to have higher total phenols.

Strong Negative Correlation: On the other hand, Flavanoids and Nonflavanoid_phenols have a correlation coefficient of -0.597872. This indicates a strong negative relationship. Wines with higher flavanoid content tend to have lower nonflavanoid phenols.

Weak Correlation: For instance, Ash and Malic_acid have a correlation coefficient of 0.183716, indicating a weak positive correlation. There is a slight tendency for wines with higher ash content to also have higher malic acid content, but the relationship is not strong.
